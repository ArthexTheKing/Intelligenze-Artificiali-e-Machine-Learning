```python
class Nodo:
    next: 'Nodo'
    prev: 'Nodo'

    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None
```

```python
class ListaDoppiamenteConcatenata:
    __head: Nodo
    __tail: Nodo

    def __init__(self):
        self.__head = None
        self.__tail = None

    def inserimento_decrescente(self, nuovo_nodo: Nodo):
        if self.__head == None:
            self.__head = nuovo_nodo
            self.__tail = nuovo_nodo
        elif nuovo_nodo.data > self.__head.data:
            nuovo_nodo.next = self.__head
            self.__head.prev = nuovo_nodo
            self.__head = nuovo_nodo
        elif nuovo_nodo.data < self.__tail.data:
            nuovo_nodo.prev = self.__tail
            self.__tail.next = nuovo_nodo
            self.__tail = nuovo_nodo
        else:
            n = self.__head
            while n.next != None and nuovo_nodo.data < n.next.data:
                if nuovo_nodo.data == n.data:
                    print('Elemento gia presente in lista')
                    return
                else:
                    n = n.next
            nuovo_nodo.next = n.next
            n.next.prev = nuovo_nodo
            n.next = nuovo_nodo
            nuovo_nodo.prev = n
    
    def cancella_elemento(self, elemento):
        trovato = False
        if self.__head == None:
            print('Lista Vuota')
            return
        elif self.__head.data == elemento:
            self.__head = self.__head.next
            self.__head.prev = None
            trovato = True
        elif self.__tail.data == elemento:
            self.__tail = self.__tail.prev
            self.__tail.next = None
            trovato = True
        else:
            n = self.__head
            while n.next != None:
                if n.data == elemento:
                    n.prev.next = n.next
                    n.next.prev = n.prev
                    trovato = True
                    break
                else:
                    n = n.next
        
        if not trovato:
            print('Elemento non presente nella lista')
```
